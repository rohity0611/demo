```python
# Import the TensorFlow library
import tensorflow as tf

# Create constant tensors with integer values
a = tf.constant(5)  # 'a' is a tensor holding the constant value 5
b = tf.constant(3)  # 'b' is a tensor holding the constant value 3

# Perform addition operation on tensors
c = a + b  # TensorFlow automatically performs element-wise addition

# Print the result using .numpy() to convert tensor to a standard Python integer
print("Addition result:", c.numpy())
```

```python
import tensorflow as tf

# Take user input for two numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

# Create constant tensors from user input
a = tf.constant(num1, dtype=tf.float32)
b = tf.constant(num2, dtype=tf.float32)

# Perform addition
c = tf.add(a, b)  # TensorFlow function for addition

# Convert tensor result to a normal Python number and print the result
print("Addition result:", c.numpy())
```

```python
import tensorflow as tf

# Take user input
num = int(input("Enter a number: "))

# Create a tensor for the input number
num_tensor = tf.constant(num, dtype=tf.int32)

# Compute remainder when divided by 2
remainder = tf.math.mod(num_tensor, 2)

# Check if the number is even or odd using TensorFlow condition
result = tf.cond(tf.equal(remainder, 0),
                 lambda: tf.constant("Even"),
                 lambda: tf.constant("Odd"))

# Print the result
print(f"The number {num} is {result.numpy()}.")
```

```python
import tensorflow as tf

# Take user input
num = float(input("Enter a number: "))

# Create a tensor for the input number
num_tensor = tf.constant(num, dtype=tf.float32)

# Define conditions
is_zero = tf.equal(num_tensor, 0)   # Check if the number is zero
is_positive = tf.greater(num_tensor, 0)  # Check if the number is positive

# Use tf.cond() to determine the category
result = tf.cond(is_zero,
                 lambda: tf.constant("Zero"),
                 lambda: tf.cond(is_positive,
                                 lambda: tf.constant("Positive"),
                                 lambda: tf.constant("Negative")))

# Print the result
print(f"The number {num} is {result.numpy()}.")
```

```python
import tensorflow as tf

# Create a TensorFlow loop using tf.range()
for num in tf.range(1, 11):
    print(num.numpy())  # Convert tensor to Python integer and print
```

```python
import tensorflow as tf

# Create a TensorFlow variable to act as a counter
counter = tf.Variable(1)

# Define the condition for the loop
def condition(counter):
    return tf.less_equal(counter, 10)  # Continue loop while counter ≤ 10

# Define the loop body
def loop_body(counter):
    print(counter.numpy())  # Convert tensor to Python integer and print
    return counter + 1  # Increment counter

# Execute the while loop
counter = tf.while_loop(condition, loop_body, [counter])
```

```python
import tensorflow as tf

# Use tf.range() to generate numbers from 10 to 1 in reverse order
for num in tf.range(10, 0, -1):  # Start at 10, stop at 1, decrement by -1
    print(num.numpy())  # Convert tensor to Python integer and print
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Synthetic Data
np.random.seed(42)  # Ensures reproducibility of random values
X = np.random.rand(100, 1) * 10  # Generate 100 random values scaled between 0 and 10
Y = 3 * X + 2 + np.random.randn(100, 1) * 2  # Define a linear relation with noise added

# Step 2: Define the Linear Regression Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation=None)  # Single neuron with linear activation (default)
])

# Step 3: Compile the Model
# - Optimizer: Stochastic Gradient Descent (SGD) - adjusts model weights to minimize error
# - Loss Function: Mean Squared Error (MSE) - measures the average squared difference between actual and predicted values
model.compile(optimizer='sgd', loss='mse')

# Step 4: Train the Model
# - X: Input feature (independent variable)
# - Y: Target variable (dependent variable)
# - epochs: Number of times the model will go through the entire dataset
# - verbose=1: Displays training progress
model.fit(X, Y, epochs=200, verbose=1)  # Train model over 200 iterations

# Step 5: Make Predictions
# Generate test input values evenly spaced between 0 and 10
X_test = np.linspace(0, 10, 100).reshape(-1, 1)  # Reshape to match model input shape
Y_pred = model.predict(X_test)  # Model predicts Y values based on trained weights

# Step 6: Visualize Results
plt.scatter(X, Y, label="Actual Data")  # Scatter plot of actual data points
plt.plot(X_test, Y_pred, color='red', label="Predicted Line")  # Red line represents model prediction
plt.xlabel("X")  # Label for X-axis
plt.ylabel("Y")  # Label for Y-axis
plt.legend()  # Show legend to differentiate actual vs predicted data
plt.title("Linear Regression using TensorFlow")  # Set plot title
plt.show()  # Display the plot
```
